```python
from sipm.analysis.SipmCalibration import SipmCalibration
import yaml
import numpy as np
```

```python
positions = ['top','bottom']
voltages = [63,65,67,69,71]
channels = [0,1,2,3]
metadata_dict = {}
for pos in positions:
    metadata_dict[pos] = {}
    for ch in channels:
        metadata_dict[pos][ch] = {}
        for volt in voltages:
            metadata_dict[pos][ch][volt] = {'metadata':{
                'date':'2022-11-22',
                'volt': volt,
                'pos': pos,
                'light': 'laser',
                'coinc': 'laser',
                'cond': 'calibration',
                'run': '*',
                'ch': ch
            }}
```

```python
print(yaml.dump(metadata_dict))
```

```python
a = SipmCalibration(
    positions=positions,
    voltages=voltages,
    channels=channels,
    directory='/scratch/gpfs/as111/results/',
    metadata_dict=metadata_dict,
    wf=False,
    merge=True,
    verbose=True
)
```

```python
a.load_files()
```

```python
bsl_thre_dict = {}
for pos in positions:
    bsl_thre_dict[pos] = {}
    for ch in channels:
        bsl_thre_dict[pos][ch] = {}
        for volt in voltages:
            bsl_thre_dict[pos][ch][volt] = 2.5
```

```python
print(yaml.dump(bsl_thre_dict))
```

```python
a.baseline_analysis(bsl_thre_dict)
```
