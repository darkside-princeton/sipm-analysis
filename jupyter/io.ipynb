```python
import matplotlib.pyplot as plt
import glob
import numpy as np
import pandas as pd
```

```python
### Please change to your own folder
path = "/scratch/gpfs/as111/results/2022-11-22/"
```

```python
data = {}
for volt in [63,65,67,69,71]:
    data[volt] = []
    files = glob.glob(f"{path}/*volt_{volt}*.h5")
    for f in files:
        df = pd.read_hdf(f, key=f'{volt}/0')
        data[volt].append(df)
    data[volt] = pd.concat(data[volt], ignore_index=True)
```

```python
for volt in [63,65,67,69,71]:
    plt.hist(data[volt]['amplitude'], bins=np.linspace(0,50,1000), histtype='step', label=f"{volt}V")
plt.legend(loc='upper right')
plt.show()
```
